```python
def sdStandardQF(
    targetVar, maxSelectorSize, groupSizeReward, centroid, selectors, data
):
    target = ps.NumericTarget(targetVar)
    searchSpace = ps.create_selectors(data, ignore=selectors)
    task = ps.SubgroupDiscoveryTask(
        data,
        target,
        searchSpace,
        result_set_size=10,
        depth=maxSelectorSize,
        qf=ps.StandardQFNumeric(a=groupSizeReward, centroid=centroid),
    )
    return ps.BeamSearch().execute(task)
```

```python
def sdStandardQFNumericTscore(targetVar, maxSelectorSize, selectors, data):
    target = ps.NumericTarget(targetVar)
    searchSpace = ps.create_selectors(data, ignore=selectors)
    task = ps.SubgroupDiscoveryTask(
        data,
        target,
        searchSpace,
        result_set_size=10,
        depth=maxSelectorSize,
        qf=ps.StandardQFNumericTscore(),
    )
    return ps.BeamSearch().execute(task)
```

### **Descrições com Todas as Variáveis**

#### Standard Quality Function for numeric targets.
    
This quality function computes interestingness of subgroups based on
the difference between subgroup mean (or median) and dataset mean (or median),
weighted by the size of the subgroup raised to the power of 'a'.

Attributes:
- a (float): Exponent to trade off between subgroup size and difference in means.
- invert (bool): Whether to invert the quality function (not used currently).
- estimator (str): Strategy for optimistic estimation ('sum', 'max', 'order').
- centroid (str): Central tendency measure ('mean', 'median', 'sorted_median').

```python
# SD 1: alvo numérico - todos os seletores - regras grandes (8) + punição 1.0 - média

result = sdStandardQF(
    targetVar="PURCHASES_TRX",
    maxSelectorSize=8,
    groupSizeReward=1.0,
    centroid='mean',
    selectors=ALL_SELECTORS,
    data=creditData,
)
result.to_dataframe()
```

**Sobre os Valores de "a"**

Ao caminhar pelos valores de a na direção de 1.0 -> 0.0, subgrupos de tamanho menor e com diferenças mais extremas na média em relação à população (para compensar a menor recompensação pelo seu pequeno tamanho) são retornados.

Subgrupos menores são descritos por um número maior de variáveis, o que dificulta a interpretação.
